```python
import requests
import numpy as np
import pandas as pd 
from bs4 import BeautifulSoup
import re
```

```python
listUrl = []
for i in range(1, 101):
    url = 'https://nettruyenco.vn/tim-truyen?status=-1&sort=10&page={}'.format(i)
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")
    listItemRow = soup.find_all("div", {"class": "row"})[2]
    listItem = listItemRow.find_all("div", {"class": "item"})
    for item in listItem:
        listUrl.append(item.find('a').get('href'))

len(listUrl)
```

```python
listHtmlComic = []
for urlComic in listUrl:
    listHtmlComic.append(requests.get(urlComic).text)

```

```python
ListTitle = []; ListAuthor = []; ListStatus = []; ListCategory = []
ListViews = []; ListRatingValue = []; ListRatingCount = []
ListFollowers = []; ListContent = []; ListChapters = []; ListComments = []
count = 0
for htmlComic in listHtmlComic:
    soupComic = BeautifulSoup(htmlComic, "html.parser")

    title = soupComic.find("h1", {"class": "title-detail"})
    if title is None:
        continue
    ListTitle.append(title.text)
    count += 1
    print(title, ' ', count)

    listInfo = soupComic.find_all("p", {"class": "col-xs-8"})
    if len(listInfo) == 3:
        author, status, category = [info.text for info in listInfo]
    else:
        author, status, category, views = [info.text for info in listInfo]
    category = category.strip()
    ListAuthor.append(author)
    ListStatus.append(status)
    ListCategory.append(category)
    ListViews.append(views)

    rating = soupComic.find("span", {"itemprop": "aggregateRating"}).find_all("span")
    ratingValue, ratingCount = rating[0].text, rating[2].text
    ListRatingValue.append(ratingValue)
    ListRatingCount.append(ratingCount)

    followers = soupComic.find("div", {"class": "follow"}).find_all("span")[2].find("b").text
    ListFollowers.append(followers)

    content = soupComic.find("div", {"class": "detail-content"}).find('p').text
    ListContent.append(content)

    chapters = soupComic.find_all("div", {"class": "col-xs-5 chapter"})[0].text.strip()
    ListChapters.append(re.findall(r'\d+', chapters)[0])

    comments = soupComic.find("span", {"class": "comment-count"}).text
    ListComments.append(comments)

df = pd.DataFrame({"Title": ListTitle, "Author": ListAuthor,
                   "Status": ListStatus, "Category": ListCategory,
                   "Views": ListViews, "Rating Value": ListRatingValue,
                   "Rating Count": ListRatingCount, "Followers": ListFollowers,
                   "Content": ListContent, "Chapters": ListChapters,
                   "Comments": ListComments})

df
```

```python
df.to_csv('./data/comic.csv', index=False)
```
